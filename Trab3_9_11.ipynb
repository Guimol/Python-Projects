```
from google.colab import files
graphfile = files.upload()
M = nx.read_pajek("/content/Days.net")
```

```
print("Grafo é direcionado?", nx.is_directed(M))
print("Grafo é ponderado?", nx.is_weighted(M))

print("Número de nós:", nx.number_of_nodes(M))
print("Número de arestas:", nx.number_of_edges(M))
```

```
print(nx.info(M))
```

```
maxNum = -1
minNum = 9999
med = 0

#degree usa peso  
for i in range(nx.number_of_nodes(M)):
  aux = M.degree(list(M.nodes)[i], weight = "weight")
  med = aux + med
  if aux < minNum:
    minNum = aux
    min = list(M.nodes)[i]
  if aux > maxNum:
    maxNum = aux
    max = list(M.nodes)[i]

med = med / nx.number_of_nodes(M)

print("Grau Máximo:", max, "-", maxNum)
print("Média dos Pesos:", med)
print("Grau Mínimo:", min, "-", minNum)
```

```
G = nx.Graph(M) #adaptação do multigrafo para grafo, com perda de informações

print(nx.info(M))
print(nx.info(G))

#nx.triangles(M) não implementado em multigrafo
nx.triangles(G)
```

```
#nx.average_clustering(M, weight="weight") não implementado em multigrafo
nx.average_clustering(G, weight="weight")
```

```
if nx.is_connected(M):
  nx.eccentricity(M)
  nx.diameter(M)
```

```
print("Número de componentes conexos:", nx.number_connected_components(M))
```

```
aux = nx.degree_centrality(M)
list1 = sorted(aux, key = aux.get, reverse = True)[:10]
list2 = list(aux.values())
list2.sort(reverse = True)
list2 = list2[:10]

important_nodes = dict(zip(list1, list2))
print(important_nodes)
```
